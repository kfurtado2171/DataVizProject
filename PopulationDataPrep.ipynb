```python
import pandas as pd
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_columns', None)
```

```python
# Importing population data
pop_data = pd.read_csv("/content/Population Estimates - US, States, Counties.csv")
pop_data.head(5)
```

```python
#Splitting the data by relevant year
pop_data_2016 = pop_data[pop_data['Year'] == 2016].copy()
pop_data_2017 = pop_data[pop_data['Year'] == 2017].copy()
pop_data_2018 = pop_data[pop_data['Year'] == 2018].copy()
pop_data_2019 = pop_data[pop_data['Year'] == 2019].copy()
pop_data_2020 = pop_data[pop_data['Year'] == 2020].copy()
```

```python
# Removing states that we dont need
state_pop_data_2016 = pop_data_2016[pop_data_2016['Countyfips'] == 0].copy()
state_pop_data_2016 = state_pop_data_2016[state_pop_data_2016['Description'] != 'U.S.'].copy()
state_pop_data_2016 = state_pop_data_2016[state_pop_data_2016['Description'] != 'Alaska'].copy()
state_pop_data_2016 = state_pop_data_2016[state_pop_data_2016['Description'] != 'Hawaii'].copy()
state_pop_data_2016.head(5)
```

```python
# Removing states that we dont need
state_pop_data_2019 = pop_data_2019[pop_data_2019['Countyfips'] == 0].copy()
state_pop_data_2019 = state_pop_data_2019[state_pop_data_2019['Description'] != 'U.S.'].copy()
state_pop_data_2019 = state_pop_data_2019[state_pop_data_2019['Description'] != 'Alaska'].copy()
state_pop_data_2019 = state_pop_data_2019[state_pop_data_2019['Description'] != 'Hawaii'].copy()
state_pop_data_2019.head(5)
```

```python
# Removing states that we dont need
state_pop_data_2020 = pop_data_2020[pop_data_2020['Countyfips'] == 0].copy()
state_pop_data_2020 = state_pop_data_2020[state_pop_data_2020['Description'] != 'U.S.'].copy()
state_pop_data_2020 = state_pop_data_2020[state_pop_data_2020['Description'] != 'Alaska'].copy()
state_pop_data_2020 = state_pop_data_2020[state_pop_data_2020['Description'] != 'Hawaii'].copy()
state_pop_data_2020.head(5)
```

```python
# Dividing by 100k to know what number we need to divide by accidents by (accidents per 100k people)
state_pop_data_2016['Population'] = state_pop_data_2016['Population'] / 100000
state_pop_data_2016.head(5)
```

```python
# Dividing by 100k to know what number we need to divide by accidents by (accidents per 100k people)
state_pop_data_2017['Population'] = state_pop_data_2017['Population'] / 100000
state_pop_data_2018['Population'] = state_pop_data_2018['Population'] / 100000
state_pop_data_2019['Population'] = state_pop_data_2019['Population'] / 100000
state_pop_data_2020['Population'] = state_pop_data_2020['Population'] / 100000
```

```python
# Getting data for all counties per year
county_pop_data_2016 = pop_data_2016[pop_data_2016['Countyfips'] != 0].copy()
county_pop_data_2017 = pop_data_2017[pop_data_2017['Countyfips'] != 0].copy()
county_pop_data_2018 = pop_data_2018[pop_data_2018['Countyfips'] != 0].copy()
county_pop_data_2019 = pop_data_2019[pop_data_2019['Countyfips'] != 0].copy()
county_pop_data_2020 = pop_data_2020[pop_data_2020['Countyfips'] != 0].copy()
```

```python
# Removing alaska and hawaii
county_pop_data_2016 = county_pop_data_2016[county_pop_data_2016['Statefips'] != 2].copy()
county_pop_data_2016 = county_pop_data_2016[county_pop_data_2016['Statefips'] != 15].copy()

county_pop_data_2017 = county_pop_data_2017[county_pop_data_2017['Statefips'] != 2].copy()
county_pop_data_2017 = county_pop_data_2017[county_pop_data_2017['Statefips'] != 15].copy()

county_pop_data_2018 = county_pop_data_2018[county_pop_data_2018['Statefips'] != 2].copy()
county_pop_data_2018 = county_pop_data_2018[county_pop_data_2018['Statefips'] != 15].copy()

county_pop_data_2019 = county_pop_data_2019[county_pop_data_2019['Statefips'] != 2].copy()
county_pop_data_2019 = county_pop_data_2019[county_pop_data_2019['Statefips'] != 15].copy()

county_pop_data_2020 = county_pop_data_2020[county_pop_data_2020['Statefips'] != 2].copy()
county_pop_data_2020 = county_pop_data_2020[county_pop_data_2020['Statefips'] != 15].copy()
```
